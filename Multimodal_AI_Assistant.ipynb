```python
import gradio as gr
from transformers import pipeline, BitsAndBytesConfig
import torch
from azure.ai.formrecognizer import DocumentAnalysisClient
from azure.core.credentials import AzureKeyCredential
import spacy
import requests
from bs4 import BeautifulSoup
from PIL import Image
```

```python
nlp = spacy.load("en_core_web_sm")

# Azure credentials and endpoint
endpoint = "you_endpoint"
key = "your_key"

# Bits and Bytes configuration for transformers
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model_id = "llava-hf/llava-1.5-7b-hf"
pipe = pipeline("image-to-text", model=model_id, model_kwargs={"quantization_config": quantization_config})
```

```python
# Function to perform OCR using Azure Form Recognizer
def perform_ocr(image_path):
    client = DocumentAnalysisClient(endpoint=endpoint, credential=AzureKeyCredential(key))
    poller = client.begin_analyze_document_from_url("prebuilt-read", image_path)
    result = poller.result()

    text = ""
    for page in result.pages:
        for line in page.lines:
            text += line.content + " "
    return text

# Function to extract key concepts using Spacy
def extract_key_concepts(text):
    doc = nlp(text)
    key_concepts = set()

    for token in doc:
        if token.pos_ in ['NOUN', 'PROPN', 'ADJ'] and not token.is_stop:
            key_concepts.add(token.text)

    for entity in doc.ents:
        key_concepts.add(entity.text)

    return key_concepts

# Function to search for resources based on key concepts
def search_resources(query):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    search_url = f"https://www.google.com/search?q={query}"
    response = requests.get(search_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    links = []
    for g in soup.find_all('div', class_='tF2Cxc'):
        anchors = g.find_all('a')
        if anchors:
            link = anchors[0]['href']
            title = g.find('h3').text if g.find('h3') else 'No title'
            snippet = g.find('span', class_='aCOpRe').text if g.find('span', class_='aCOpRe') else 'No snippet'
            links.append({"title": title, "url": link, "snippet": snippet})

    return links

# Function to analyze the image and generate a description
def analyze_image(image_url):
    # Open and resize image
    image = Image.open(requests.get(image_url, stream=True).raw).resize((750, 500))

    # Generate image description using the model pipeline
    prompt = "USER: <image>\nWhat is this picture\nASSISTANT: "
    outputs = pipe(image, prompt=prompt, generate_kwargs={"max_new_tokens": 200})
    image_description = outputs[0]['generated_text']

    return image_description

# Main function to process the image URL
def process_image(image_url):
    # Analyze image to get a description
    image_description = analyze_image(image_url)

    # Perform OCR to extract text from the image
    ocr_text = perform_ocr(image_url)
    combined_text = image_description + " " + ocr_text
    print("Extracted Text:", combined_text)

    # Extract key concepts from the combined text
    key_concepts = extract_key_concepts(combined_text)
    print("Extracted Key Concepts:", key_concepts)

    # Combine all key concepts into a single search query
    combined_query = " ".join(key_concepts)
    search_results = search_resources(combined_query)

    # Format resource links
    resource_links = []
    for result in search_results:
        resource_links.append({
            "title": result['title'],
            "url": result['url'],
            "snippet": result['snippet']
        })

    return image_description, resource_links

```

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from transformers import pipeline, T5Tokenizer, T5ForConditionalGeneration
from langchain.base_language import BaseLanguageModel
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.schema import LLMResult, Generation
import torch
from torch.quantization import quantize_dynamic
from typing import Optional, List, Dict, Any
from pydantic import BaseModel, Field, PrivateAttr
```

```python

class HuggingFaceLLM(BaseLanguageModel):
    model_name: str
    model: T5ForConditionalGeneration = PrivateAttr()
    tokenizer: T5Tokenizer = PrivateAttr()
    pipeline: Any = PrivateAttr()

    def __init__(self, model_name: str, **data):
        super().__init__(model_name=model_name, **data)
        self.model_name = model_name
        self.model = T5ForConditionalGeneration.from_pretrained(model_name)
        self.tokenizer = T5Tokenizer.from_pretrained(model_name)
        self.model = quantize_dynamic(self.model, {torch.nn.Linear}, dtype=torch.qint8)
        self.pipeline = pipeline('summarization', model=self.model, tokenizer=self.tokenizer, device=-1, num_beams=5, early_stopping=True)

    def _call(self, prompt: str, stop: Optional[List[str]] = None, **kwargs) -> LLMResult:
        inputs = self.tokenizer(prompt, return_tensors="pt", truncation=True, padding=True, max_length=self.model.config.max_length)
        input_ids = inputs["input_ids"]
        if input_ids.shape[1] > self.model.config.max_length:
            raise ValueError("Input text is too long for the model's max position embeddings")
        response = self.pipeline(prompt, max_length=200, min_length=50, num_beams=5, early_stopping=True)
        generations = [Generation(text=r['summary_text']) for r in response]
        return LLMResult(generations=[generations])

    async def _acall(self, prompt: str, stop: Optional[List[str]] = None, **kwargs) -> LLMResult:
        return self._call(prompt, stop, **kwargs)

    def generate_prompt(self, prompt: str, stop: Optional[List[str]] = None, **kwargs) -> LLMResult:
        return self._call(prompt, stop, **kwargs)

    async def agenerate_prompt(self, prompt: str, stop: Optional[List[str]] = None, **kwargs) -> LLMResult:
        return self._acall(prompt, stop, **kwargs)

    def predict(self, input_text: str, stop: Optional[List[str]] = None, **kwargs) -> str:
        return self._call(input_text, stop, **kwargs).generations[0].text

    async def apredict(self, input_text: str, stop: Optional[List[str]] = None, **kwargs) -> str:
        return (await self._acall(input_text, stop, **kwargs)).generations[0].text

    def predict_messages(self, messages: List[Dict[str, Any]], stop: Optional[List[str]] = None, **kwargs) -> str:
        combined_input = " ".join([msg["content"] for msg in messages])
        return self.predict(combined_input, stop, **kwargs)

    async def apredict_messages(self, messages: List[Dict[str, Any]], stop: Optional[List[str]] = None, **kwargs) -> str:
        combined_input = " ".join([msg["content"] for msg in messages])
        return await self.apredict(combined_input, stop, **kwargs)

    def invoke(self, prompt: str, **kwargs) -> LLMResult:
        return self._call(prompt, **kwargs)

    class Config:
        arbitrary_types_allowed = True

# Define the chunk text function
def chunk_text(text: str, chunk_size: int, overlap: int = 50) -> List[str]:
    words = text.split()
    chunks = [" ".join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size - overlap)]
    return chunks

# Define the summarization function
def summarize_pdf(pdf_file):
    loader = PyPDFLoader(pdf_file.name)
    documents = loader.load()
    if not documents:
        raise ValueError("No documents were loaded. Please check the document source.")

    text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200)
    texts = text_splitter.split_documents(documents)

    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
    persist_directory = "/tmp/chromadb"
    vectordb = Chroma.from_documents(documents=texts, embedding=embeddings, persist_directory=persist_directory)
    vectordb.persist()

    prompt_template = """Forget whatever summary was generated and Write a concise summary of the following: "{context}" CONCISE SUMMARY: """
    prompt = PromptTemplate(template=prompt_template, input_variables=["context"])

    fine_tuned_model_path = "/content/drive/MyDrive/fine_tuned_model"
    llm = HuggingFaceLLM(model_name=fine_tuned_model_path)

    qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=vectordb.as_retriever(), chain_type="stuff", chain_type_kwargs={"prompt": prompt}, return_source_documents=True)

    query = "please summarize this book"
    results = vectordb.as_retriever().get_relevant_documents(query)
    if not results:
        raise ValueError("No relevant documents were retrieved for the query.")

    context = " ".join([doc.page_content for doc in results])

    chunk_size = llm.tokenizer.model_max_length - 50
    chunks = chunk_text(context, chunk_size, overlap=50)

    summaries = []
    for chunk in chunks:
        response = llm._call(chunk)
        summary_text = response.generations[0][0].text if response.generations else ""
        summaries.append(summary_text)

    final_summary = " ".join(summaries)
    user_friendly_summary = final_summary.replace('\n', ' ').replace('  ', ' ')

    return user_friendly_summary
```

```python
def image_analysis_interface(image_url):
    image_description, resource_links = process_image(image_url)
    resource_links_formatted = "\n\n".join([f"[{link['title']}]({link['url']})\n{link['snippet']}" for link in resource_links])
    return image_description, resource_links_formatted

def pdf_summary_interface(pdf_file):
    summary = summarize_pdf(pdf_file)
    return summary

# Create the Gradio interface
with gr.Blocks() as demo:
    with gr.Tab("Image Analyzer"):
        image_url_input = gr.Textbox(label="Image URL")
        image_description_output = gr.Textbox(label="Image Description")
        resource_links_output = gr.Markdown(label="Resource Links")
        analyze_image_button = gr.Button("Analyze Image")
        analyze_image_button.click(image_analysis_interface, inputs=[image_url_input], outputs=[image_description_output, resource_links_output])

    with gr.Tab("PDF Summarizer"):
        pdf_file_input = gr.File(label="Upload PDF")
        pdf_summary_output = gr.Textbox(label="PDF Summary")
        summarize_pdf_button = gr.Button("Summarize PDF")
        summarize_pdf_button.click(pdf_summary_interface, inputs=[pdf_file_input], outputs=[pdf_summary_output])
```
